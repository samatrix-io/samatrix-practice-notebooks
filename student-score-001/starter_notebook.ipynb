# Samatrix AIML Practice: Student Score Prediction

Use the training data to build a regression model and generate `predictions.csv`.

Samatrix scores the CSV. It does not execute this notebook in the MVP.


```python
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
sample = pd.read_csv('sample_submission.csv')

train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

features = ['study_hours', 'attendance_rate', 'previous_score', 'assignment_score']
target = 'final_score'

model = RandomForestRegressor(random_state=42, n_estimators=100)
model.fit(train[features], train[target])

predictions = model.predict(test[features])

submission = pd.DataFrame({
    'id': test['id'],
    'prediction': predictions
})

submission.to_csv('predictions.csv', index=False)
submission.head()
```
